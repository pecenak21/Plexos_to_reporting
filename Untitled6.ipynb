```python
"""
convert.py
Handles the native PLEXOS Cloud solution extraction using the PLEXOS CLI tool via subprocess.
"""
import os
import shutil
import subprocess
from pathlib import Path

def find_plexos_cli():
    """
    Dynamically locates the plexos-cloud CLI utility without hardcoded user paths.
    Checks system PATH first, then scans standard default installation folders.
    """
    # 1. Try to find it in the environment's current PATH variables
    cli_match = shutil.which("plexos-cloud")
    if cli_match:
        return cli_match
        
    # 2. Fallback: Search the common system-wide program file paths for Energy Exemplar
    fallback_directories = [
        Path(os.environ.get("ProgramFiles", "C:\\Program Files")) / "Energy Exemplar",
        Path(os.environ.get("ProgramFiles(x86)", "C:\\Program Files (x86)")) / "Energy Exemplar",
        Path(os.environ.get("LocalAppData", "")) / "Programs" / "Energy Exemplar"
    ]
    
    for base_dir in fallback_directories:
        if base_dir.exists():
            # Recursively find plexos-cloud.exe or plexos-cloud.cmd within the Energy Exemplar folder
            for extension in ["*.exe", "*.cmd", "*.bat"]:
                matches = list(base_dir.rglob(f"plexos-cloud{extension[1:]}"))
                if matches:
                    return str(matches[0])
                    
    return None

def convert_zip_to_parquet(zip_file_path):
    """
    Checks if an extraction directory already exists. If not, runs the native command:
    plexos-cloud solution convert zip-to-parquet --zipFile <path_to_zip> -d <output_directory>
    
    Returns:
        Path object to the extracted parquet directory, or None if it fails.
    """
    zip_path = Path(zip_file_path).resolve()
    
    # Validation checks
    if not zip_path.exists():
        print(f"[-] Error: Target zip file does not exist at '{zip_path}'")
        return None
        
    # Derive output directory from zip name
    output_dir = zip_path.parent / zip_path.stem
    
    # Light check: skip if the directory already exists and isn't empty
    if output_dir.exists() and any(output_dir.iterdir()):
        print(f"[~] Directory folder '{output_dir.name}' already exists. Skipping native PLEXOS CLI conversion.")
        return output_dir
        
    # Discover the tool location dynamically
    plexos_cli_path = find_plexos_cli()
    if not plexos_cli_path:
        print("[-] Error: 'plexos-cloud' CLI utility could not be located in your PATH or standard install paths.")
        print("[-] Please ensure PLEXOS Cloud CLI is installed on this machine.")
        return None
        
    print(f"[+] Dynamic location discovered: {plexos_cli_path}")
    print(f"[+] Creating target directory: '{output_dir}'")
    output_dir.mkdir(parents=True, exist_ok=True)
    
    # Construct the command line arguments
    cmd = [
        f'"{plexos_cli_path}"', "solution", "convert", "zip-to-parquet",
        "--zipFile", f'"{str(zip_path)}"',
        "-d", f'"{str(output_dir)}"'
    ]
    
    cmd_string = " ".join(cmd)
    print(f"[+] Executing native PLEXOS command: {cmd_string}")
    
    try:
        result = subprocess.run(cmd_string, check=True, capture_output=True, text=True, shell=True)
        print("[+] PLEXOS CLI conversion completed successfully.")
        return output_dir
        
    except subprocess.CalledProcessError as e:
        print(f"[-] Error: PLEXOS native conversion failed.")
        print(f"[-] CLI StdErr output:\n{e.stderr}")
        print(f"[-] CLI StdOut output:\n{e.stdout}")
        return None
    except Exception as e:
        print(f"[-] Unexpected error tracking the CLI process: {e}")
        return None
```

```python
"""
database.py
Features case-insensitive, whitespace-tolerant, and None-safe SQL queries over local Parquet structures.
"""
import os
import re
import duckdb
import pandas as pd

def initialize_database_structures(base_dir):
    try:
        duckdb.query("SELECT 1 FROM mem_fki LIMIT 1")
    except duckdb.CatalogException:
        print("[+] Building high-performance metadata cache...")
        base_dir_clean = str(base_dir).replace('\\', '/')
        
        data_path = f"{base_dir_clean}/data/**/*.parquet"
        fki_path = f"{base_dir_clean}/fullkeyinfo/**/*.parquet"
        period_path = f"{base_dir_clean}/period/**/*.parquet"
        
        duckdb.query(f"CREATE TEMPORARY TABLE mem_fki AS SELECT * FROM read_parquet('{fki_path}')")
        duckdb.query(f"CREATE TEMPORARY TABLE mem_period AS SELECT * FROM read_parquet('{period_path}')")
        duckdb.query(f"CREATE VIEW v_data AS SELECT * FROM read_parquet('{data_path}')")

def get_automatic_scale_factor(base_dir, property_name, target_header):
    initialize_database_structures(base_dir)
    lookup_prop = property_name[0] if isinstance(property_name, list) else property_name
    
    query = f"SELECT unitValue FROM mem_fki WHERE LOWER(TRIM(propertyName)) = '{lookup_prop.lower().strip()}' LIMIT 1;"
    row = duckdb.query(query).fetchone()
    db_unit = row[0].strip() if row else ""
    
    matches = re.findall(r'\(([^)]+)\)', target_header)
    target_unit = matches[-1].strip() if matches else ""
    
    if not db_unit or not target_unit:
        return 1.0, True
    db_clean, target_clean = db_unit.lower().replace(" ", ""), target_unit.lower().replace(" ", "")
    if db_clean == target_clean:
        return 1.0, True
    
    return 1.0, True

def pull_pivoted_data(base_dir, property_name, unique_months, category_list=None, class_name=None, emission_gas_name=None, temporal_pattern="monthly"):
    """
    Executes a high-performance database PIVOT with case-insensitive and trailing-whitespace tolerant filters.
    """
    initialize_database_structures(base_dir)
    
    props = property_name if isinstance(property_name, list) else [property_name]
    classes = class_name if isinstance(class_name, list) else [class_name] * len(props)
    
    if isinstance(category_list, list) and len(category_list) > 0 and isinstance(category_list[0], list):
        cats = category_list
    else:
        cats = [category_list] if not isinstance(category_list, list) else [category_list] * len(props)

    if emission_gas_name:
        props = ['Generation Production']
        classes = [None]
        cats = [None]

    filter_blocks = []
    for p, c, clist in zip(props, classes, cats):
        sub_conditions = [f"LOWER(TRIM(f.PropertyName)) = '{p.lower().strip()}'"]
        if c:
            sub_conditions.append(f"LOWER(TRIM(f.childClassName)) = '{c.lower().strip()}'")
            
        if clist:
            valid_cats = [cat for cat in clist if cat is not None and str(cat).lower().strip() != 'all']
            if valid_cats:
                placeholders = ", ".join(f"'{str(cat).lower().strip()}'" for cat in valid_cats)
                sub_conditions.append(f"LOWER(TRIM(f.ChildObjectCategoryName)) IN ({placeholders})")
        
        filter_blocks.append("(" + " AND ".join(sub_conditions) + ")")
        
    master_filter = " OR ".join(filter_blocks)
    gas_filter = f"AND LOWER(TRIM(f.parentObjectName)) = '{emission_gas_name.lower().strip()}'" if emission_gas_name else ""

    if temporal_pattern == "daily":
        day_select = "CAST(EXTRACT(DAY FROM CAST(p.StartDate AS TIMESTAMP)) AS INTEGER) AS Day_Id,"
        group_by_clause = "GROUP BY Object_Name, band_id, Year, Day_Id"
    else:
        day_select = ""
        group_by_clause = "GROUP BY Object_Name, band_id, Year"

    # Fast logic to grab the proper name (Parent for Zones, Child for Gens)
    # If a list of properties is passed, append the property name so pandas stacks them instead of squashing them.
    is_list_mode = isinstance(property_name, list) and len(property_name) > 1
    if is_list_mode:
        obj_name_expr = "(CASE WHEN f.childObjectName IS NULL OR TRIM(f.childObjectName) = '' THEN f.parentObjectName ELSE f.childObjectName END) || ' - ' || f.PropertyName"
    else:
        obj_name_expr = "(CASE WHEN f.childObjectName IS NULL OR TRIM(f.childObjectName) = '' THEN f.parentObjectName ELSE f.childObjectName END)"

    pivot_query = f"""
        PIVOT (
            SELECT 
                {obj_name_expr} AS Object_Name,
                f.bandId AS band_id,
                CAST(EXTRACT(YEAR FROM CAST(p.StartDate AS TIMESTAMP)) AS INTEGER) AS Year,
                {day_select}
                strftime(CAST(p.StartDate AS TIMESTAMP), '%b-%Y') AS Month_Label,
                d.Value AS Metric_Value
            FROM v_data d
            JOIN mem_fki f ON d.SeriesId = f.seriesId AND d.DataFileId = f.dataFileId
            JOIN mem_period p ON d.PeriodId = p.PeriodId
            WHERE ({master_filter})
              {gas_filter}
        )
        ON Month_Label
        USING SUM(Metric_Value)
        {group_by_clause}
    """
    
    try:
        df = duckdb.query(pivot_query).df()
        if df.empty: 
            return df
            
        # Fill sparse NaN intersections with 0.0 so pandas grouping engines can slice safely
        meta_cols = ['Object_Name', 'band_id', 'Year', 'Day_Id']
        pivot_cols = [c for c in df.columns if c not in meta_cols]
        df[pivot_cols] = df[pivot_cols].fillna(0.0)
        
        # Guarantee presence of all missing horizon columns
        for m in unique_months:
            if m not in df.columns:
                df[m] = 0.0
                
        return df
    except Exception:
        return pd.DataFrame()
```

```python
"""
transformers.py
Processes pivoted queries into standardized matrix visuals, handling flat, banded, and daily granularities.
"""
import pandas as pd
import numpy as np
# from config import BAND_MAP, BANDS_ORDERED
# from database import pull_pivoted_data, get_automatic_scale_factor

def process_daily_block(parquet_base_dir, property_name, header_name, years, unique_months, is_rate=False, category_list=None, class_name=None):
    """
    Generates isolated 1-31 row sections per unique resource asset name.
    """
    df_pivoted = pull_pivoted_data(
        parquet_base_dir, property_name, unique_months, 
        category_list=category_list, class_name=class_name, temporal_pattern="daily"
    )
    if df_pivoted.empty: return pd.DataFrame()
    
    scale_factor, _ = get_automatic_scale_factor(parquet_base_dir, property_name, header_name)
    df_pivoted[list(unique_months)] = df_pivoted[list(unique_months)] * scale_factor
    df_grouped = df_pivoted.groupby(['Object_Name', 'Day_Id'])[list(unique_months)].sum()

    if is_rate:
        df_grid, final_cols = build_rate_totals(df_grouped, years, unique_months)
    else:
        df_grid, final_cols = build_sum_totals(df_grouped, years, unique_months)
        
    unique_assets = sorted(df_pivoted['Object_Name'].unique())
    all_chunks = []
    standard_days = list(range(1, 32))
    
    for asset in unique_assets:
        asset_slice = df_grid.loc[asset].reindex(standard_days, fill_value=0.0)
        asset_slice.index.name = ''
        asset_header = pd.DataFrame(columns=final_cols, index=[asset])
        blank_row = pd.DataFrame(columns=final_cols, index=[''])
        all_chunks.extend([asset_header, asset_slice, blank_row])
        
    return pd.concat(all_chunks) if all_chunks else pd.DataFrame()

def process_emissions_block(parquet_base_dir, gas_name, target_header, years, unique_months):
    df_pivoted = pull_pivoted_data(parquet_base_dir, 'Production', unique_months, emission_gas_name=gas_name)
    if df_pivoted.empty: return pd.DataFrame()
    df_pivoted = df_pivoted.groupby('Object_Name')[list(unique_months)].sum()
    scale_factor, _ = get_automatic_scale_factor(parquet_base_dir, 'Production', target_header)
    df_pivoted = df_pivoted * scale_factor
    df_grid, _ = build_sum_totals(df_pivoted, years, unique_months)
    df_grid.loc['Total'] = df_grid.sum(axis=0)
    df_grid.index.name = ''
    return df_grid

def process_flat_block(parquet_base_dir, property_name, header_name, years, unique_months, is_rate=False, category_list=None, class_name=None):
    df_pivoted = pull_pivoted_data(parquet_base_dir, property_name, unique_months, category_list=category_list, class_name=class_name)
    if df_pivoted.empty: return pd.DataFrame()
    df_pivoted = df_pivoted.groupby('Object_Name')[list(unique_months)].sum()
    scale_factor, _ = get_automatic_scale_factor(parquet_base_dir, property_name, header_name)
    df_pivoted = df_pivoted * scale_factor
    if is_rate:
        df_grid, _ = build_rate_totals(df_pivoted, years, unique_months)
        df_grid.loc['Total'] = df_grid.replace(0, np.nan).mean(axis=0).fillna(0)
    else:
        df_grid, _ = build_sum_totals(df_pivoted, years, unique_months)
        df_grid.loc['Total'] = df_grid.sum(axis=0)
    df_grid.index.name = ''
    return df_grid

def process_banded_block(parquet_base_dir, property_name, header_name, years, unique_months, category_list=None, class_name=None):
    df_pivoted = pull_pivoted_data(parquet_base_dir, property_name, unique_months, category_list=category_list, class_name=class_name)
    if df_pivoted.empty: return pd.DataFrame()
    df_pivoted['Band_Label'] = df_pivoted['band_id'].map(BAND_MAP).fillna('  (Other)')
    scale_factor, _ = get_automatic_scale_factor(parquet_base_dir, property_name, header_name)
    df_pivoted[list(unique_months)] = df_pivoted[list(unique_months)] * scale_factor
    df_pivoted = df_pivoted.set_index(['Object_Name', 'Band_Label'])
    df_grid, cols = build_sum_totals(df_pivoted[list(unique_months)], years, unique_months)
    
    flat_rows = []
    unique_generators = sorted(df_pivoted.index.get_level_values('Object_Name').unique())
    for gen in unique_generators:
        gen_data = df_grid.loc[gen]
        flat_rows.append([gen] + list(gen_data.sum(axis=0)))
        for band in BANDS_ORDERED:
            flat_rows.append([band] + list(gen_data.loc[band]) if band in gen_data.index else [band] + [0.0]*len(cols))
                
    df_final = pd.DataFrame(flat_rows, columns=['Asset_Label'] + cols)
    bottom_sums = df_grid.reset_index().drop(columns=['Object_Name']).groupby('Band_Label').sum()
    df_final.loc[len(df_final)] = ['Total'] + list(df_grid.sum(axis=0))
    for band in BANDS_ORDERED:
        df_final.loc[len(df_final)] = [band] + list(bottom_sums.loc[band]) if band in bottom_sums.index else [band] + [0.0]*len(cols)
    return df_final

def build_sum_totals(df_pivot, years, unique_months):
    final_columns = []
    grand_total_series = pd.Series(0.0, index=df_pivot.index)
    for yr in years:
        yr_months = [m for m in unique_months if f"-{yr}" in m]
        final_columns.extend(yr_months)
        yr_total_col = f"Tot-{yr}"
        df_pivot[yr_total_col] = df_pivot[yr_months].sum(axis=1)
        final_columns.append(yr_total_col)
        grand_total_series += df_pivot[yr_total_col]
    df_pivot['Total'] = grand_total_series
    final_columns.append('Total')
    return df_pivot[final_columns], final_columns

def build_rate_totals(df_pivot, years, unique_months):
    final_columns = []
    for yr in years:
        yr_months = [m for m in unique_months if f"-{yr}" in m]
        final_columns.extend(yr_months)
        yr_total_col = f"Tot-{yr}"
        df_pivot[yr_total_col] = df_pivot[yr_months].replace(0, np.nan).mean(axis=1).fillna(0)
        final_columns.append(yr_total_col)
    df_pivot['Total'] = df_pivot[unique_months].replace(0, np.nan).mean(axis=1).fillna(0)
    final_columns.append('Total')
    return df_pivot[final_columns], final_columns

def build_combined_emissions_section(parquet_base_dir, active_gases, topline_header, years, unique_months):
    combined_rows = []
    for gas in active_gases:
        sub_header = f"Total Effluents (lb) -- {gas}"
        df_gas = process_emissions_block(parquet_base_dir, gas, topline_header, years, unique_months)
        if not df_gas.empty:
            subheader_row = pd.DataFrame([[None] * len(df_gas.columns)], index=[sub_header], columns=df_gas.columns)
            combined_rows.append(subheader_row)
            combined_rows.append(df_gas)
            combined_rows.append(pd.DataFrame([[None] * len(df_gas.columns)], index=[''], columns=df_gas.columns))
    return pd.concat(combined_rows) if combined_rows else pd.DataFrame()
```

```python
"""
run_report.py
Orchestrates conversion execution and runs master report loops across composite property structures.
"""
import os
import pandas as pd
import duckdb
# from config import REPORT_BLUEPRINT, ASSET_GROUPS
# from convert import convert_zip_to_parquet
# from transformers import (
#     process_flat_block, 
#     process_banded_block,
#     process_daily_block,
#     build_combined_emissions_section
# )

def execute_pipeline(zip_file_path, output_csv):
    parquet_base_dir = convert_zip_to_parquet(zip_file_path)
    if parquet_base_dir is None: return
    
    base_dir_clean = str(parquet_base_dir).replace('\\', '/')
    period_path = f"{base_dir_clean}/period/**/*.parquet"
    
    print(f"[+] Verifying baseline time dimensions from path: {parquet_base_dir}")
    time_df = duckdb.query(f"""
        SELECT DISTINCT 
            CAST(EXTRACT(YEAR FROM CAST(StartDate AS TIMESTAMP)) AS INTEGER) AS Year,
            strftime(CAST(StartDate AS TIMESTAMP), '%b-%Y') AS Month_Label,
            CAST(StartDate AS TIMESTAMP) as raw_date
        FROM read_parquet('{period_path}')
        ORDER BY raw_date
    """).df()
    
    years = sorted(time_df['Year'].unique())
    unique_months = time_df['Month_Label'].unique()
    
    compiled_blocks = []
    print("[+] Beginning single master report sequence...")
    
    for header, class_input, group_input, prop_input, temp_pattern, is_banded, is_rate in REPORT_BLUEPRINT:
        print(f"    -> Processing Section: {header}")
        
        is_list_mode = isinstance(prop_input, list)
        if is_list_mode:
            target_categories = [None if gk == "all" else ASSET_GROUPS.get(gk, None) for gk in group_input]
        else:
            target_categories = None if group_input == "all" else ASSET_GROUPS.get(group_input, None)
            
        first_class = class_input[0] if is_list_mode else class_input
        
        if first_class == "Emission":
            df_block = build_combined_emissions_section(
                parquet_base_dir, ["CO2", "NOx", "SO2"], header, years, unique_months
            )
            idx_flag = True
        elif temp_pattern == "daily":
            df_block = process_daily_block(
                parquet_base_dir, prop_input, header, years, unique_months,
                is_rate=is_rate, category_list=target_categories, class_name=class_input
            )
            idx_flag = True
        elif is_banded:
            df_block = process_banded_block(
                parquet_base_dir, prop_input, header, years, unique_months, 
                category_list=target_categories, class_name=class_input
            )
            idx_flag = False
        else:
            df_block = process_flat_block(
                parquet_base_dir, prop_input, header, years, unique_months, 
                is_rate=is_rate, category_list=target_categories, class_name=class_input
            )
            idx_flag = True

        compiled_blocks.append((header, str(prop_input), df_block, idx_flag))
        
    print(f"[+] Streaming master sequential compilation to: {output_csv}")
    with open(output_csv, 'w', newline='', encoding='utf-8') as f:
        for header, prop_name, df_block, idx_flag in compiled_blocks:
            f.write(f"{header}\n")
            if df_block is not None and not df_block.empty:
                df_block.to_csv(f, index=idx_flag, header=True, lineterminator='\n')
                f.write("\n\n")
            elif df_block is None:
                f.write("\n")
            else:
                f.write(f"Data not found for properties matching sequence\n\n\n")
            
    print("[+] Report generation complete!")

if __name__ == "__main__":
    ZIP_FILE_INPUT = r"C:\Users\cto\Downloads\Model TA_Base2_ST Solution.zip"
    OUTPUT_REPORT = r"C:\Users\cto\Downloads\std_report.csv"
    execute_pipeline(ZIP_FILE_INPUT, OUTPUT_REPORT)
```
